#### Phân chia lại dataset : 70% train - 20% validate - 10% test

```python
import os
import random
import shutil
from collections import defaultdict

# Đường dẫn gốc
dataset_path = '/content/datasets'
images_path = os.path.join(dataset_path, 'images')
labels_path = os.path.join(dataset_path, 'labels')

# Tạo cấu trúc thư mục mới
for split in ['train', 'val', 'test']:
    os.makedirs(os.path.join(dataset_path, split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(dataset_path, split, 'labels'), exist_ok=True)

# 1. Phân loại ảnh theo Class dựa trên file nhãn
class_groups = defaultdict(list)
all_images = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

for img_f in all_images:
    label_f = img_f.rsplit('.', 1)[0] + '.txt'
    label_p = os.path.join(labels_path, label_f)

    if os.path.exists(label_p):
        with open(label_p, 'r') as f:
            lines = f.readlines()
            if lines:
                # Lấy class ID của đối tượng đầu tiên trong ảnh
                cls_id = lines[0].split()[0]
                class_groups[cls_id].append(img_f)
            else:
                class_groups['empty'].append(img_f)
    else:
        class_groups['no_label'].append(img_f)

# 2. Chia riêng từng Class và gộp lại
train_files, val_files, test_files = [], [], []

for cls_id, files in class_groups.items():
    random.shuffle(files)
    n = len(files)
    idx1 = int(n * 0.7)
    idx2 = idx1 + int(n * 0.2)

    train_files.extend(files[:idx1])
    val_files.extend(files[idx1:idx2])
    test_files.extend(files[idx2:])
    print(f"Class {cls_id}: Total={n}, Train={idx1}, Val={idx2-idx1}, Test={n-idx2}")

# 3. Di chuyển file
def move_files(files, split):
    for f in files:
        shutil.move(os.path.join(images_path, f), os.path.join(dataset_path, split, 'images', f))
        label_f = f.rsplit('.', 1)[0] + '.txt'
        if os.path.exists(os.path.join(labels_path, label_f)):
            shutil.move(os.path.join(labels_path, label_f), os.path.join(dataset_path, split, 'labels', label_f))

move_files(train_files, 'train')
move_files(val_files, 'val')
move_files(test_files, 'test')

print(f"\n Đã chia xong: {len(train_files)} train, {len(val_files)} val, {len(test_files)} test.")
```

## Bước 3: Tạo File Cấu Hình và Script Huấn Luyện

```python
import torch
import ultralytics.utils.loss as loss_module
from ultralytics import YOLO

class SoftBCEWithLogitsLoss(torch.nn.Module):
    def __init__(self, original_bce, sigma=15.0):
        super().__init__()
        self.original_bce = original_bce
        self.sigma = sigma
        self.print_counter = 0

    def forward(self, pred_scores, target_scores):
        nc = target_scores.shape[-1]
        device = target_scores.device
        dtype = target_scores.dtype

        fg_mask = target_scores.sum(dim=-1) > 0
        soft_target_scores = target_scores.clone()

        if fg_mask.sum() > 0:
            pos_targets = target_scores[fg_mask]
            assigned_cls = pos_targets.argmax(dim=-1)
            original_iou_scores = pos_targets.sum(dim=-1, keepdim=True)

            class_angles = torch.arange(nc, device=device, dtype=dtype) * 15.0
            true_angles = class_angles[assigned_cls]

            diff = (true_angles[:, None] - class_angles[None, :]).abs()
            d = torch.minimum(diff, 360.0 - diff)

            soft = torch.exp(-0.5 * (d / self.sigma) ** 2)
            soft = soft * original_iou_scores
            soft_target_scores[fg_mask] = soft

            self.print_counter += 1
            if self.print_counter % 50 == 1:
                print("\n" + "="*70)
                print("[DEBUG] SO SÁNH TARGET TRƯỚC VÀ SAU KHI LÀM MỀM (Soft Target)")
                print("Original target_scores (Hard):")
                print(target_scores[fg_mask][0].cpu().numpy().round(3))
                print("\nModified target_scores (Soft - Sigma = {}):".format(self.sigma))
                print(soft_target_scores[fg_mask][0].cpu().numpy().round(3))
                print("="*70 + "\n")

        return self.original_bce(pred_scores, soft_target_scores)

original_init = loss_module.v8DetectionLoss.__init__
def patched_init(self, model):
    original_init(self, model)
    print("\n>>> Đã kích hoạt thành công Soft Angular Target (BCE Loss) với Sigma=15.0 <<<\n")
    self.bce = SoftBCEWithLogitsLoss(self.bce, sigma=15.0)

loss_module.v8DetectionLoss.__init__ = patched_init

```

```python
# 1. Create train.py with SoftBCE Loss
train_py_content = """
import argparse
import torch
import ultralytics.utils.loss as loss_module
from ultralytics import YOLO

class SoftBCEWithLogitsLoss(torch.nn.Module):
    def __init__(self, original_bce, sigma=15.0):
        super().__init__()
        self.original_bce = original_bce
        self.sigma = sigma
        self.print_counter = 0

    def forward(self, pred_scores, target_scores):
        nc = target_scores.shape[-1]
        device = target_scores.device
        dtype = target_scores.dtype

        fg_mask = target_scores.sum(dim=-1) > 0
        soft_target_scores = target_scores.clone()

        if fg_mask.sum() > 0:
            pos_targets = target_scores[fg_mask].float()
            assigned_cls = pos_targets.argmax(dim=-1)
            original_iou_scores = pos_targets.sum(dim=-1, keepdim=True)

            class_angles = torch.arange(nc, device=device, dtype=torch.float32) * 15.0
            true_angles = class_angles[assigned_cls]

            diff = (true_angles[:, None] - class_angles[None, :]).abs()
            d = torch.minimum(diff, 360.0 - diff)

            soft = torch.exp(-0.5 * (d / self.sigma) ** 2)
            soft = soft * original_iou_scores

            soft_target_scores[fg_mask] = soft.to(dtype)

            self.print_counter += 1
            if self.print_counter % 50 == 1:
                orig = target_scores[fg_mask][0].cpu().numpy().round(3)
                soft_tgt = soft_target_scores[fg_mask][0].cpu().numpy().round(3)
                print(f"\\n[SoftBCE] Orig: {orig}\\n[SoftBCE] Soft: {soft_tgt}\\n")

        return self.original_bce(pred_scores, soft_target_scores)

original_init = loss_module.v8DetectionLoss.__init__
def patched_init(self, model):
    original_init(self, model)
    print(f"\\n[SoftBCE] Activated (Sigma=15.0)")
    self.bce = SoftBCEWithLogitsLoss(self.bce, sigma=15.0)

loss_module.v8DetectionLoss.__init__ = patched_init

def parse_args():
    parser = argparse.ArgumentParser()
    parser.add_argument('--data', default='leanbot_data.yaml')
    parser.add_argument('--epochs', type=int, default=100)
    parser.add_argument('--batch', type=int, default=16)
    parser.add_argument('--name', default='leanbot_colab')
    return parser.parse_args()

if __name__ == '__main__':
    args = parse_args()
    model = YOLO("yolo11n.pt")
    model.train(
        data=args.data,
        epochs=args.epochs,
        batch=args.batch,
        device=0,
        name=args.name,
        degrees=10.0, fliplr=0.0, flipud=0.0
    )
"""
with open('train.py', 'w', encoding='utf-8') as f:
    f.write(train_py_content.strip())

# 2. Create leanbot_data.yaml
yaml_content = """
path: /content/datasets
train: train/images
val: val/images
test: test/images
nc: 24
names:
  0: Leanbot_0
  1: Leanbot_p15
  2: Leanbot_p30
  3: Leanbot_p45
  4: Leanbot_p60
  5: Leanbot_p75
  6: Leanbot_p90
  7: Leanbot_p105
  8: Leanbot_p120
  9: Leanbot_p135
  10: Leanbot_p150
  11: Leanbot_p165
  12: Leanbot_p180
  13: Leanbot_p195
  14: Leanbot_m150
  15: Leanbot_m135
  16: Leanbot_m120
  17: Leanbot_m105
  18: Leanbot_m90
  19: Leanbot_m75
  20: Leanbot_m60
  21: Leanbot_m45
  22: Leanbot_m30
  23: Leanbot_m15
"""
with open('leanbot_data.yaml', 'w', encoding='utf-8') as f:
    f.write(yaml_content.strip())
```

## Bước 6: Chạy thử Inference (Dự đoán) trên ảnh mẫu


```python
from ultralytics import YOLO
import glob
import cv2
from google.colab.patches import cv2_imshow
import os


model_path = '/content/runs/detect/leanbot_colab/weights/best.pt'

if os.path.exists(model_path):
    model = YOLO(model_path)
    # Lấy ảnh từ tập test để chạy thử dự đoán
    sample_images = glob.glob('/content/datasets/test/images/*.jpg')

    if sample_images:
        print(f" Đang chạy dự đoán với model: {model_path}")
        # Giữ conf=0.25 vì mAP của bạn rất cao, không cần giảm quá thấp
        results = model.predict(source=sample_images, conf=0.25)
        for result in results:
            res_plotted = result.plot()
            cv2_imshow(res_plotted)
            print(f"Kết quả cho: {os.path.basename(result.path)}")
    else:
        print(" Không tìm thấy ảnh trong thư mục /content/datasets/test/images/")
else:
    print(f" Không tìm thấy model tại: {model_path}")
```
